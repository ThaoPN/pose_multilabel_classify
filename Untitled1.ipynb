```python
from keras.preprocessing.image import img_to_array
from keras.models import load_model
import numpy as np
import imutils
import pickle
import cv2
import os
```

```python
def prediction(file_path, model, mlb):
    image = cv2.imread(file_path)

    # pre-process the image for classification
    image = cv2.resize(image, (96, 96))
    image = image.astype("float") / 255.0
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)

#     model = load_model('pose_multi_labels.h5')
#     mlb = pickle.loads(open('mlb.pickle', "rb").read())

    proba = model.predict(image)[0]
    idxs = np.argsort(proba)[::-1][:2]
    print(idxs)
    # build the label and draw the label on the image
    label = "{}_{}".format(mlb.classes_[idxs[0]], mlb.classes_[idxs[1]])
    
    return label

# print(prediction(image_path))
```

```python
model = load_model('pose_multi_labels.h5')
mlb = pickle.loads(open('mlb.pickle', "rb").read())

def get_files_path(path, file_extension):
    file_content = []
    for root, dirs, files in os.walk(path):
        for file in files:
            lower_file = file.lower()
            if lower_file.endswith(file_extension):
                file_path = os.path.join(path, file)
                file_content.append(file_path)
        break
    return file_content

files = get_files_path('./dataset/test_set/bending_h', '.jpg')

for f in files:
    print(prediction(f, model, mlb))
```

```python
files = get_files_path('./dataset/test_set/others_h', '.jpg')
print(len(files))
for f in files:
    print(prediction(f, model, mlb))
```

```python
image_path = './dataset/test_set/bending_h/13_20180502093101_0090_0.jpg'

image = cv2.imread(image_path)

# pre-process the image for classification
image = cv2.resize(image, (96, 96))
image = image.astype("float") / 255.0
image = img_to_array(image)
image = np.expand_dims(image, axis=0)

model = load_model('pose_multi_labels.h5')
mlb = pickle.loads(open('mlb.pickle', "rb").read())

proba = model.predict(image)[0]
idxs = np.argsort(proba)[::-1][:2]

print(idxs)
```

```python
np.set_printoptions(suppress=True)
print(proba)
# loop over the indexes of the high confidence class labels
for (i, j) in enumerate(idxs):
    # build the label and draw the label on the image
    label = "{}: {:.2f}%".format(mlb.classes_[j], proba[j] * 100)
    print(label)
```

```python
# show the probabilities for each of the individual labels
for (label, p) in zip(mlb.classes_, proba):
    print("{}: {:.2f}%".format(label, p * 100))
```
